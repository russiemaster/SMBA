```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7 , test_size = 0.3,
random_state = 0)
print(df_train.shape)
print(df_test.shape)
```

```python
df_train['depressed'].value_counts()
sns.countplot(x='depressed', data=df_train, palette='flare')
plt.show()
```

# **Model 1**

```python
from imblearn.over_sampling import SMOTE
import statsmodels.api as sm
os = SMOTE(random_state=0)
df_train_x=df_train[['Married','sex','Age','education_level','Number_children']]
df_train_x = sm.add_constant(df_train_x)
df_train_y=df_train[['depressed']]
columns = df_train_x.columns
os_df_X,os_df_y=os.fit_resample(df_train_x, df_train_y)
os_df_X = pd.DataFrame(data=os_df_X,columns=columns )
os_df_y= pd.DataFrame(data=os_df_y,columns=['depressed'])
X=os_df_X[columns]
y=os_df_y['depressed']
import statsmodels.api as sm
logit_model=sm.Logit(y,X)
result=logit_model.fit()
print(result.summary2())


from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(df_train_x, df_train_y)
df_test_x=df_test[['Married','sex','Age','education_level','Number_children']]
df_test_x = sm.add_constant(df_test_x)
df_test_y=df_test[['depressed']]
y_pred = logreg.predict(df_test_x)
print('Accuracy of logistic regression classifier on test set:{:.2f}'.format(logreg.score(df_test_x, df_test_y)))

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(df_test_y, logreg.predict(df_test_x))
fpr, tpr, thresholds = roc_curve(df_test_y, logreg.predict_proba(df_test_x)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
